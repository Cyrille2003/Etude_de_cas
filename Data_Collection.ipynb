```python
import numpy as np
import pandas as pd
from classe_reacteur import Reacteur
```

```python
from combinaisons import combis, liste_reacs
def gen_data():
    data = []
    for combinaison in combis:
        moment_cree = np.array([0,0,0])
        force_cree = np.array([0,0,0])
        for i, actif in enumerate(combinaison):
            if actif == 1:
                moment_cree = np.add(moment_cree, liste_reacs[i].moment_cree())
                force_cree = np.add(force_cree, liste_reacs[i].force_cree())
        
        data.append(
            pd.DataFrame([{"Combinaison":combinaison, 
                                     "Moment créé [Nm]":moment_cree, 
                                     "Force créée":force_cree}])
                    )
    data = pd.concat(data, ignore_index=True)

    return data

data = gen_data()
print(data)
```
